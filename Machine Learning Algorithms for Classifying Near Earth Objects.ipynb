## The Data

There are plenty objects including asteroids, meteors, satellites and space debris from human activities or space exploration found in space, these objects can either be hazardous or non-hazardous to  the Earth. These objects may seem so far away from us but when looked at on a grand scale, they are relatively close to the Earth and there is a possibility that they can have a significant impact on Earth. The asteroids are being monitored and recorded by The National Aeronautics and Space Administration. This dataset contains several features of the objects being tracked and they are referred to as Near-Earth Objects.

The dataset is from Kaggle, an online platform where individuals with a data science background contribute datasets and resources for users. The specific data used in this project has been produced by Sameep Vani, combining the National Aeronautics and Space Administration Application Programming Interface (NASA API) and NEO Earth close approaches from the California institute of technology. The dataset has two datafiles neo.csv and neo_v2.csv, the dataset that has been selected is neo_v2,  the newest uploaded version.

Each asteroid is given an identification number, they are also named by The National Aeronautics and Space Administration. The estimated size of the asteroids are stated in the dataset, the estimated minimum and maximum diameter of the asteroids in kilometers, which is important for understanding the possible effect of the object. Information about the object's relative velocity (km/h) to Earth is also captured in the dataset along with the miss distance in kilometers, the trajectory of the asteroids can be calculated and possible threats can be assessed. The orbiting body specifies the planet around which the object revolves. The sentry object variable is an automated collision monitoring system for assessing possible future impacts of the object on Earth. The absolute magnitude column describes the intrinsic brightness of the asteroid. The hazardous column indicates whether the object poses a threat, which is important for the Earth’s safety assessment.

## Clustering and K-Means Model

```python
# A subset is made and to be used for clustering; it can be altered to test different variables, here the miss distance and absolute magnitudes are used.
clustering_subset = nasa_df[["Miss Distance km", "Absolute Magnitude (M)"]]
```

The function below finds the number of clusters to be used in the K-means model:

```python
def clustering_K(subset):

        distortions = [] #create an empty list for the distortion data
        K = range(1,10) #the range of K values for the number of clusters to be tested
        for k in K: #for each value in K
            kmeanModel = KMeans(n_clusters=k) #build a K-Means clustering model
            kmeanModel.fit(subset) #fit the model to the data
            distortions.append(kmeanModel.inertia_) #add the results to the list of distortions 

        plt.figure(figsize=(6,5))
        plt.plot(K, distortions, 'bx-')
        plt.xlabel('k')
        plt.ylabel('Distortion')
        plt.title('The Elbow Method showing the optimal k')
        plt.show()
```

The next function then performs the K-Means Model:

```python
def K_means(number_of_clusters, subset):

        kmeans = KMeans(n_clusters = number_of_clusters)
        kmeans.fit(subset)
        kmeans_labels = kmeans.labels_

        silhouette_score = metrics.silhouette_score(subset, kmeans_labels)
        silhouette_score_rounded = round(silhouette_score, 3)

        print(f"The Silhouette score is {silhouette_score_rounded}")

        subset['kClusters'] = kmeans_labels
        print(subset['kClusters'].value_counts())

        x_index = 0 
        y_index = 1  

        fig = sns.relplot(data = subset, x = subset.iloc[:, x_index], y = subset.iloc[:, y_index], s = 100, hue = "kClusters", palette = "Paired") 

        plt.xlabel(subset.columns[x_index]) 
        plt.ylabel(subset.columns[y_index]) 
        plt.title("K Means Clusters")
        plt.show()
```

## Receiver Operating Characteristic (ROC) Curve

Roc curve is useful in python for showing the performance of the classification methods in predicting the true positive outcome or negative outcome 

```python
def roc_curve_func(y_test, y_scores, method):
    fpr, tpr, threshold = roc_curve(y_test, y_scores[:, 1])
    plt.title('Receiver Operating Characteristic')
    roc_auc = auc(fpr, tpr)
    plt.plot(fpr, tpr, 'b', label='AUC = %0.2f' % roc_auc)
    plt.legend(loc='center right')
    plt.plot([0, 1], [0, 1], 'r--')
    plt.xlim([0, 1])
    plt.ylim([0, 1])
    plt.ylabel('True Positive Rate')
    plt.xlabel('False Positive Rate')
    plt.title('ROC Curve of ' + method)
    plt.show()
    return roc_auc
```

_This definition uses the true test set and the predicted scores for the class 1 and the name of the classifier method depending on what is being used. It calculates a false positive (fpr), a true positive (tpr) and the threshold for roc curves, creating the area under the ROC curve._
_The code calls on the function roc_curve_fun using the predicted  scores, and the methods._

## Function to Compute the Confusion Matrix:

A confusion matrix for Knn is another way of displaying how many accurate and inaccurate predictions there are for class 0 and 1

```python
def confusion_matrix_print(Y_Test, review_preditctions):
    confusion_matrix = metrics.confusion_matrix(Y_Test, review_preditctions)
    plt.matshow(confusion_matrix)
    plt.title('Confusion Matrix')
    plt.colorbar()
    plt.ylabel('True class')
    plt.xlabel('Predicted class')
    plt.show()
    return 
```

## 2. Neural Network

Define a function for a Neural Network with its parameters

```python
def NeuralNetwork(x_train, y_train, x_test, y_test,Comparison_Table, Print_Graphs):
    
    start = time.time()## Note the start time to calculate the excecution time
    
    ## Standard scaler is used to standardize the training set and testing set   
    scaler = StandardScaler()
    x_train = scaler.fit_transform(x_train)
    x_test = scaler.transform(x_test)
    
    ## Define a sequential Neural Network model    
    model = keras.Sequential([
    layers.Dense(64, activation='relu', input_dim=x_train.shape[1]),
    layers.Dense(32, activation='relu'),
    layers.Dense(1, activation='sigmoid') 
    ])

    ## Compile the Neural Network model
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    ## Train the Neural Network model on the training data   
    history = model.fit(x_train, y_train, epochs=10, batch_size=32, validation_data=(x_train, y_train))
    ## Evaluate the model on the test set
    loss, accuracy = model.evaluate(x_test, y_test)
    print(f'Test Loss: {loss:.4f}') 
    print(f'Test Accuracy: {accuracy:.4f}')
    
    ## Predict the probilities of the test set 
    y_pred = model.predict(x_test) ## Create a dataframe to store the predicted values
    reviews_pred = pd.DataFrame(list(y_pred)) ## Convert the predicted values to binary
    y_pred_binary = (y_pred > 0.5).astype(int)
    
    ## Calculate the recall score, precision score, f1 score 
    recall = recall_score(y_test, y_pred_binary)
    precision = precision_score(y_test, y_pred_binary)
    f1 = f1_score(y_test, y_pred_binary)    
   
    ## Calcultate the total run time of the function  
    timeTaken = time.time() - start
    
    ## Create a new row with recall score, precision score, f1 score and the excecution time 
    new_row= pd.DataFrame({'Machine_Learning_Algorithm': "Neural Network",
                           'Recall': recall, 'Precision':precision,
                           'F1_Score':f1, "Run_time":timeTaken }, index=[0])
    
    ## Add the new row to the table created earelier
    Comparison_Table = pd.concat([new_row,Comparison_Table.loc[:]]).reset_index(drop=True)
    
    if Print_Graphs ==True:
        ## Print the confusion matrix 
        confusion_matrix_print(y_test, y_pred_binary)
    
        # Plot the training and validation accuracy over epochs
        plt.plot(history.history['accuracy'], label='Training Accuracy')
        plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
        plt.xlabel('Epoch')
        plt.ylabel('Accuracy')
        plt.legend()
        plt.show()
    
    ## Return the updated table
    return Comparison_Table
    
```

```python
def Naive_Bayes(x_train,y_train,x_test,y_test,Table, Print_Graphs):
    
    start= time.time()
    NB = GaussianNB()
    NB.fit(x_train,y_train)
    score = NB.score(x_test, y_test)
    Naive_Bayes_score_rounded = round(score, 3)
   
    predictions = NB.predict(x_test)
    reviews_pred = pd.DataFrame(list(predictions))

    confusion_matrix = metrics.confusion_matrix(y_test, reviews_pred)
    
    timeTaken = time.time() - start
    new_row= pd.DataFrame({'Machine_Learning_Algorithm': "Naive Bayes", 'Recall':recall_score(y_test,reviews_pred), 'Precision':precision_score(y_test,reviews_pred), 'F1_Score':f1_score(y_test,reviews_pred) ,"Run_time":timeTaken }, index=[0])
    Table = pd.concat([new_row,Table.loc[:]]).reset_index(drop=True)
    
    if Print_Graphs == True:
        
        confusion_matrix_print(y_test, reviews_pred)
    
        y_scores_pred = NB.predict_proba(x_test)
        auc_knn = roc_curve_func(y_test, y_scores_pred, 'Naive Bayes')
    
   
    return Table

```

## 6. Decision Tree

```python
def Decision_Tree(x_train, y_train, x_test, y_test, Table, Print_Graphs):
    ## The diffrent types of decision tree
    Criteria_Array= ["gini", "log_loss", "entropy"]
    ## Looping through the diffrent types 
    for i in range(0,3):
        ## Calculate execution time
        start = time.time();
        ## DT object 
        DT = DecisionTreeClassifier(criterion = Criteria_Array[i], max_depth= 3, random_state=0)
        ##Training 
        DT.fit(x_train, y_train)
        ##Testing 
        y_pred = DT.predict(x_test)
        timeTaken = time.time() - start
        
        if Print_Graphs == True:
            ## Visulation 
            fig = plt.figure(figsize=(25,20))
            print(Criteria_Array[i])
            tree.plot_tree(DT, feature_names=["Est Diameter Min km","Est Diameter Max km","'Relative Velocity km/h","Miss Distance km","Absolute Magnitude (M)"], class_names=['non hazardous', 'hazardous'], filled=True);
        
            
            #Confusion Matrix 
            confusion_matrix_print(y_test, y_pred)
            
            ## ROC Curve 
            y_scores_pred = DT.predict_proba(x_test)
            auc_knn = roc_curve_func(y_test, y_scores_pred, Criteria_Array[i])
            
            
            
            
        ## Table of Results 
        new_row= pd.DataFrame({'Machine_Learning_Algorithm': Criteria_Array[i], 'Recall':recall_score(y_test,y_pred), 'Precision':precision_score(y_test,y_pred), 'F1_Score':f1_score(y_test,y_pred) ,"Run_time":timeTaken }, index=[0])
        Table = pd.concat([new_row,Table.loc[:]]).reset_index(drop=True)
       
        
    return Table
```
